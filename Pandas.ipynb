```python
Option1 ####### Reading CSV File without PANDAS#########
import csv
def calculate_ratings_stats(data, industry=None):
    ratings = []
    for row in data:
        if row[3]!='NULL' and (not industry or row[1]==industry):
            ratings.append(float(row[3]))
    max_rating= max(ratings)
    min_rating= min(ratings)
    avg_rating= sum(ratings)/len(ratings)
    return max_rating, min_rating, avg_rating
    
with open(r"C:\Users\Public\Documents\self\Technical\DataAnalysiss\Maths & Statistics\Pandas\movies.csv") as f:
    data = list(csv.reader(f))
    header = data[0]
    data = data[1:]
    #print(data)

    max_rating, min_rating, avg_rating = calculate_ratings_stats(data)
    print(f"All records: Min rating={min_rating}, Max Rating = {max_rating}, Avg ratings = {avg_rating}")

    max_rating, min_rating, avg_rating = calculate_ratings_stats(data, industry="Bollywood")
    print(f"Bollywood: Min rating={min_rating}, Max Rating = {max_rating}, Avg ratings = {avg_rating}")

    max_rating, min_rating, avg_rating = calculate_ratings_stats(data, industry="Hollywood")
    print(f"Hollywood: Min rating={min_rating}, Max Rating = {max_rating}, Avg ratings = {avg_rating}")
```

```python
#define a custom function
def grouper(df, idx, col):
    if 80 <= df[col].loc[idx] <=90:
        return '80-90'
    elif 50<= df[col].loc[idx] <=60:
        return '50-60'
    else:
        return 'others'
```

```python
g= df.groupby(lambda idx: grouper(df, idx, 'temperature'))
g
```

```python
for key, d in g:
    print("key:",key)
    print("Data:",d)
```
