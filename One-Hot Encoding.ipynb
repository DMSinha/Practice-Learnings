# One Hot Encoding with Scikit Learn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh_enc = OneHotEncoder(sparse = False)
#if you want to drop any column, pass the paramter as drop=first/last
#note: when we use dummies, we get dataframe ; but when we use OneHotEncoder we get output in numpy array 
```

```python
oh_enc_arr=oh_enc.fit_transform(tips_df[['sex','smoker','day','time']])
```

```python
oh_enc_arr
```

```python
dummy_df.keys()
```

```python
#to convert this numpy array into dataframe
oh_enc_df = pd.DataFrame(oh_enc_arr,columns=['sex_Female', 'sex_Male', 'smoker_No',
       'smoker_Yes', 'day_Fri', 'day_Sat', 'day_Sun', 'day_Thur',
       'time_Dinner', 'time_Lunch'])
```

```python
oh_enc_df
#here , we can see that for all the categorical data we have, we have converted them into dummy variables
```
